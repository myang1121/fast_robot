```python
import sys
#%load_ext autoreload
#%autoreload 2

# for data collection
from ble import get_ble_controller
from base_ble import LOG
from cmd_types import CMD_lab6
import time

# for plotting
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft

LOG.propagate = False
```

```python
# Get ArtemisBLEController object
ble = get_ble_controller()

# Connect to the Artemis Device
ble.connect()
```

```python
# debugging data
T_arr = []
Yaw_arr = []
Error_arr = []
Control_speed_arr = []
```

```python
# clear old data
T_arr.clear()
Yaw_arr.clear()
Error_arr.clear()
Control_speed_arr.clear()
```

```python
def orientation_pid_data_notif_handler(uuid, byte_array):
    # get data from Artemis through ble
    s = ble.bytearray_to_string(byte_array)
    
    # split string (e.g. "T:1234|Y:45|E:5|C:60") into key-value pairs
    s_split = dict(item.split(":") for item in s.split("|"))

    # append values to arrays for plotting
    T_arr.append(int(s_split["T"]))
    Yaw_arr.append(float(s_split["Y"]))
    Error_arr.append(float(s_split["E"]))
    Control_speed_arr.append(float(s_split["C"]))
```

```python
ble.start_notify(ble.uuid['RX_STRING'], orientation_pid_data_notif_handler)
```

```python
T_arr.clear()
Yaw_arr.clear()
Error_arr.clear()
Control_speed_arr.clear()

# to avoid accumulated= drift between upload, reset yaw first
ble.send_command(CMD_lab6.RESET_YAW, "")
print("Reset Yaw to 0°")
# Change target angle to 90° while PID is running
ble.send_command(CMD_lab6.SET_ORIENTATION_SETPOINT, "0.0")
print("New setpoint 0°")
# to avoid accumulated= drift between upload, reset yaw first
ble.send_command(CMD_lab6.RESET_YAW, "")
print("Reset Yaw to 0°")
# start orientation PID controller (motors start)
ble.send_command(CMD_lab6.START_ORIENTATION_PID, "")
print("Orientation PID start!")
```

```python
# Change target angle to 90° while PID is running
ble.send_command(CMD_lab6.SET_ORIENTATION_SETPOINT, "0.0")
print("New setpoint 0°")
```

```python
# stop PID controller (motors stop)
ble.send_command(CMD_lab6.STOP_ORIENTATION_PID, "")
print("Orientation PID stop!")

# make sure the motors have completely stopped
time.sleep(0.5) 

# Artemis start BLE send data loop
ble.send_command(CMD_lab6.SEND_ORIENTATION_PID_DATA, "")
print("Requesting data...")
time.sleep(10)
print(f"Received {len(T_arr)} data points")
```

```python
# plot
T = np.array(T_arr)
T = T - T[0]

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 8), sharex=True)

# Yaw angle data vs time
ax1.plot(T, Yaw_arr, color='blue', label='Yaw angle (°)')
ax1.axhline(y=0, color='green', linestyle='--', label='Setpoint (°)')
ax1.set_ylabel('Yaw (°)')
ax1.legend(); ax1.grid(True)
ax1.set_title(f'Orientation P Controller - Rotation Angle Control ({len(T)} samples)')

# Error vs time
ax2.plot(T, Error_arr, color='red', label='Error (°)')
ax2.axhline(y=0, color='white', linestyle='--', linewidth=0.5)
ax2.axhline(y=2,  color='gray',  linestyle=':',  label='±2° deadzone')
ax2.axhline(y=-2, color='gray',  linestyle=':')
ax2.set_ylabel('Error (°)')
ax2.legend(); ax2.grid(True)

# Motor control speed vs time
ax3.plot(T, Control_speed_arr, color='orange', label='Control Speed (PWM)')
ax3.set_ylabel('PWM')
ax3.set_xlabel('Time (ms)')
ax3.legend(); ax3.grid(True)

plt.tight_layout()
plt.show()

# Final result based on controller quality
print(f"Run duration:   {T[-1]}ms")
print(f"Final yaw: {Yaw_arr[-1]:.2f}°")
print(f"Final error:    {Error_arr[-1]:.2f}°")
```
